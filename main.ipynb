```python
from tkinter import *
from tkinter import filedialog as fa
import pytesseract
from PIL import ImageTk,Image
from gtts import gTTS
import os
import pyqrcode
import qrcode
import ctypes
import webbrowser

myappid = 'mycompany.myproduct.subproduct.version' # arbitrary string
ctypes.windll.shell32.SetCurrentProcessExplicitAppUserModelID(myappid)

root = Tk()
root.geometry("360x540")
root.configure(background='#33cc33')
root.iconbitmap(r'C:\Users\amitg\Desktop\T_T\juyp\logo_2.ico')
root.title("Image_Text_Audio_Qr_Web")
file=" "
flag = 0
tag=0
qr_tag=0
aud_tag=0
web_tag=0
w=Label(root,text="Project",fg='black',padx=10)
w.grid(row=1,columnspan=2)
def op():
    global file
    file=fa.askopenfilename(title = "Choose an IMAGE")
    print(file)
    global flag 
    flag=1
    w=Label(root,text="Project",fg='black',padx=10)
    w.grid(row=1,columnspan=2)
    
def fi():
    if flag == 0:
        w=Label(root,text=" Please Select An IMAGE First :",fg='black')
        w.grid(row=1,columnspan=2)
    else:
        print(file)
        img=Image.open(file.encode("utf-8"),"r")
        
        pytesseract.pytesseract.tesseract_cmd = "C:/Program Files/Tesseract-OCR/tesseract.exe"
        txt = pytesseract.image_to_string(img)
        print(txt)
        global tag 
        tag=1
        fo=open('filey.txt','w',encoding="utf-8")
        fo.write(txt)
        print ("Converted to Text ")
        w=Label(root,text="Project",fg='black',padx=10)
        w.grid(row=1,columnspan=2)
        

def au():
    if flag ==0 & tag==0 :
        w=Label(root,text="Please Select A TXT FILE First",fg='black')
        w.grid(row=1,columnspan=2)
    else:
        lan='en'
    
        fo=open('filey.txt','r', encoding="utf-8")
        tx=fo.read()
        my=gTTS(text=tx,lang=lan,slow=False)
        
        my.save('audio.mp3')
        print("Audio converted")
        global aud_tag 
        aud_tag=1
        w=Label(root,text="Project",fg='black',padx=10)
        w.grid(row=1,columnspan=2)
        
def qr():

    if flag ==0 & tag==0:
        w=Label(root,text="Please select A TXT FILE First :",fg='black')
        w.grid(row=1,columnspan=2)
    else:
        with open('filey.txt','r', encoding="utf-8") as f:
            data = f.read()
        Logo_link = 'logo.png'
        logo = Image.open(Logo_link)
        basewidth = 100
        wpercent = (basewidth/float(logo.size[0]))
        hsize = int((float(logo.size[1])*float(wpercent)))
        logo = logo.resize((basewidth, hsize), Image.ANTIALIAS)
        QRcode = qrcode.QRCode(error_correction=qrcode.constants.ERROR_CORRECT_H)

 
        # adding URL or text to QRcode
        QRcode.add_data(data)
 
        # generating QR code
        QRcode.make()
 
        # taking color name from user
        QRcolor = 'Green'
 
        # adding color to QR code
        img = QRcode.make_image(fill_color=QRcolor, back_color="white").convert('RGB')
 
        # set size of QR code
        pos = ((img.size[0] - logo.size[0]) // 2,(img.size[1] - logo.size[1]) // 2)
        img.paste(logo, pos)
 
        # save the QR code generated
        img.save("qrcode.png")
        print("QR Generated")
        global qr_tag 
        qr_tag=1
        w=Label(root,text="Project",fg='black',padx=10)
        w.grid(row=1,columnspan=2)
        
def cw():
    if flag ==0 & tag==0:
        w=Label(root,text="Please select A TXT FILE First :",fg='black')#bg='White',fg='blue')
        w.grid(row=1,columnspan=2)
    else:    
        contents = open('filey.txt','r', encoding="utf-8")
        with open("filey.html", "w",encoding="utf-8") as e:
            for lines in contents.readlines():
                e.write("<pre>" + lines + "</pre> <br>\n")
        print("Webpage Created")
        global web_tag 
        web_tag=1
        w=Label(root,text="Project",fg='black',padx=10)
        w.grid(row=1,columnspan=2)

def opt():
    if tag==0:
        w=Label(root,text="Please Generate TXT FILE First ",fg='black')
        w.grid(row=1,columnspan=2)
    else:
        if tag==1:
            os.system("filey.txt")
            w=Label(root,text="Project",fg='black',padx=10)
            w.grid(row=1,columnspan=2)
    
    
def opa():
    if aud_tag==0:
        w=Label(root,text="Please Create An AUDIO File :",fg='black')
        w.grid(row=1,columnspan=2)  
    else:
        os.system("audio.mp3")
        w=Label(root,text="Project",fg='black',padx=10)
        w.grid(row=1,columnspan=2)
    
def opi():
    if flag ==0:
        w=Label(root,text=" Please Select An IMAGE First :",fg='black')
        w.grid(row=1,columnspan=2)  
    else:
        os.system(file)
        w=Label(root,text="Project",fg='black',padx=10)
        w.grid(row=1,columnspan=2)
    
def oqr():
    if qr_tag==0:
        w=Label(root,text="Please Generate QR Code First",fg='black')
        w.grid(row=1,columnspan=2)
    else:
        os.system("qrcode.png")
        w=Label(root,text="Project",fg='black',padx=10)
        w.grid(row=1,columnspan=2)
    
def ocw():
    if web_tag==0:
        w=Label(root,text="Please Create A Webpage First",fg='black')
        w.grid(row=1,columnspan=2)
    else:
        x=webbrowser.open_new_tab('filey.html')
        w=Label(root,text="Project",fg='black',padx=10)
        w.grid(row=1,columnspan=2)
        if x==True:
            print("Website Opened")


l=Label(root,text=" Image =>Text => Audio/Qr/Web ",padx=30,pady=20,bg='white',fg='#33cc33',width=20,font=('aerial',18))#,bg='White',fg='blue',width=20,font=('aerial',18))
b=Button(root,text='SELECT AN  IMAGE ',padx=20,pady=20,command=op,bg='#ffffff',fg='#00cc00',borderwidth=0)
b1=Button(root,text='CONVERT TO  TEXT ',padx=20,pady=20,command=fi,bg='#ffffff',fg='#00cc00',borderwidth=0)
b2=Button(root,text='CONVERT TO AUDIO',padx=16.5,pady=20,command=au,bg='#ffffff',fg='#00cc00',borderwidth=0)
b3=Button(root,text='OPEN THE TXT File',padx=18,pady=20,command=opt,bg='#ffffff',fg='#00cc00',borderwidth=0)
b4=Button(root,text='OPEN THE AUDIO',padx=20,pady=20,command=opa,bg='#ffffff',fg='#00cc00',borderwidth=0)
b5=Button(root,text='OPEN THE IMAGE',padx=20,pady=20,command=opi,bg='#ffffff',fg='#00cc00',borderwidth=0)
b6=Button(root,text='CONVERT  TO  QR',padx=24,pady=20,command=qr,bg='#ffffff',fg='#00cc00',borderwidth=0)
b7=Button(root,text='OPEN  QR Code',padx=25,pady=20,command=oqr,bg='#ffffff',fg='#00cc00',borderwidth=0)
b8=Button(root,text='CONVERT TO WEB',padx=22,pady=20,command=cw,bg='#ffffff',fg='#00cc00',borderwidth=0)
b9=Button(root,text='OPEN WEBPAGE',padx=25,pady=20,command=ocw,bg='#ffffff',fg='#00cc00',borderwidth=0)
b.grid(row =2,column=0,padx=15,pady=10,sticky=W)
b1.grid(row=3,column=0,sticky=W,padx=15,pady=10)
b2.grid(row=4,column=0,sticky=W,padx=15,pady=10)
b3.grid(row=3,column=1,sticky=W,padx=10,pady=10)
b4.grid(row=4,column=1,sticky=W,padx=10,pady=10)
b5.grid(row=2,column=1,sticky=W,padx=10,pady=10)
b6.grid(row=5,column=0,sticky=W,padx=15,pady=10)
b7.grid(row=5,column=1,sticky=W,padx=10,pady=10)
b8.grid(row=6,column=0,sticky=W,padx=15,pady=10)
b9.grid(row=6,column=1,sticky=W,padx=10,pady=10)
l.grid(row = 0,columnspan=2,padx=10,pady=10,sticky=W)


root.mainloop()
```
